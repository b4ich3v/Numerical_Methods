```python
from sympy import symbols, diff, Eq, solve
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = symbols("a")
b = symbols("b")

nodes = [0, 1, 2, 3, 4]
values = [0, 1, 1, 2, 2]
```

```python
expression = 0
for x, y in zip(nodes, values):
    expression += (y - (a * x + b)) ** 2

divA = diff(expression, a)  
divB = diff(expression, b)  

solutions = solve([Eq(divA, 0), Eq(divB, 0)], [a, b], dict = True)

a_sol = solutions[0][a]
b_sol = solutions[0][b]
```

```python
def linear_function(x):
    return a_sol * x + b_sol 

new_values = np.linspace(-1,5, 100)
plt.scatter(nodes, values)
plt.plot(new_values, linear_function(new_values), "red")
plt.show()
```

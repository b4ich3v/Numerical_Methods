```python
from sympy import symbols, diff, Eq, solve
import numpy as np
import matplotlib.pyplot as plt
```

```python
nodes = [0, 1, 2, 3, 4]
values = [1, 3, 8, 21, 60]

expression = 0
a, b = symbols("a b", real = True)
```

```python
for x, y in zip(nodes, values):
    expression += (y - (a * x + b)) ** 2

solutions = solve([Eq(diff(expression, a), 0), Eq(diff(expression, b), 0) ], [a, b], dict = True)
a_sol = solutions[0][a]
b_sol = solutions[0][b]

def lin_func(x):
    return a_sol * x + b_sol

x_range = np.linspace(1, 50, 100)
plt.scatter(nodes, values)
plt.plot(x_range, lin_func(x_range), "red")
plt.show()
```

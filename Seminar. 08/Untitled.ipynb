```python
from sympy import symbols, diff, Eq, solve
import numpy as np
import matplotlib.pyplot as plt
```

```python
nodes = [0, 1, 2, 3, 4]
values = [1, 3, 8, 21, 60]

expression = 0
alfa, b = symbols("alfa b", real = True)
```

```python
for x, y in zip(nodes, values):
    expression += (np.log(y) - (alfa + x * b)) ** 2

solutions = solve([Eq(diff(expression, alfa), 0), Eq(diff(expression, b),0)], [alfa, b], dict = True)
alfa_sol = solutions[0][alfa]
b_sol = solutions[0][b]
a_sol = np.exp(float(alfa_sol))

def exponential_func(x):
    return a_sol * np.e ** (x * b_sol)

x_range = np.linspace(1, 50, 100)
plt.scatter(nodes, values)
plt.plot(x_range, exponential_func(x_range), "red")
plt.show()
```

```python
from sympy import symbols, diff, Eq, solve
import numpy as np
import matplotlib.pyplot as plt
```

```python
nodes = [0, 1, 2, 3, 4]
values = [1, 3, 8, 21, 60]

c0, c1, c2, c3 = symbols("c0 c1 c2 c3", real = True)
```

```python
expression = 0
for x, y in zip(nodes, values):
    expression += (y - (c0 + c1 * x + c2 * x ** 2 + c3 * x ** 3)) ** 2

solutions = solve([Eq(diff(expression, c0),0), Eq(diff(expression, c1),0),
                  Eq(diff(expression, c2),0), Eq(diff(expression, c3),0)], [c0, c1, c2, c3], dict = True)

c0_sol = solutions[0][c0]
c1_sol = solutions[0][c1]
c2_sol = solutions[0][c2]
c3_sol = solutions[0][c3]

def cubic_func(x):
    return c0_sol + c1_sol * x + c2_sol * x ** 2 + c3_sol * x ** 3


x_range = np.linspace(1, 50, 100)
plt.scatter(nodes, values)
plt.plot(x_range, cubic_func(x_range), "red")
plt.show()
```

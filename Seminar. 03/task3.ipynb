```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def runge(x):
    return 1 / (1 + 25 * x**2)

size = 11
x_points = [math.cos((2 * (current + 1) - 1) / (2 * size) * math.pi) for current in range(size)]
y_points = [runge(current) for current in x_points]

def lagrange_pol(x_points, y_points, x):
    result = 0
    
    for i in range(len(x_points)):
        term = y_points[i]
        for j in range(len(y_points)):
            if i != j:
                term *= (x - x_points[j]) / (x_points[i] - x_points[j])
        result += term
    return result
```

```python
x_range = np.linspace(-1, 1, 100)
plt.plot(x_range, lagrange_pol(x_points, y_points, x_range))
plt.plot(x_range, runge(x_range))
plt.show()
```

```python
plt.plot(x_range, abs(lagrange_pol(x_points, y_points, x_range) - runge(x_range)))
plt.show()
```
